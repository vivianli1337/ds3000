```python
import random as rnd
from collections import Counter
# create n "jurors" with verdict accuracy in range from 0.5 to max
n = 11
max_accuracy = 1.0

# _ is an anonymonous variable
jury = [rnd.uniform(0.50, max_accuracy) for _ in range(n)]
jury
```

```python
# create a series of outcomes (0=innocent; 1=guilty)
trials = 100000
outcomes = [rnd.choice([0,1]) for _ in range(trials)]
print(outcomes)
```

```python
# for each outcome, each juror makes a decision
# dont require unanimity - the majority makes the decision

verdicts = []
for out in outcomes:
    verdict = [] # track invidual decision of each juror
    for juror in jury:
        if rnd.random() < juror: # generate random number from 0-1
            verdict.append(out)
        else:
            verdict.append(1-out)
    # the majority        
    verdicts.append(Counter(verdict).most_common(1)[0][0])

verdicts
```

```python
list(zip(outcomes, verdicts))
```

```python
# compute the jury majority verdict accuracy
correct = 0
for i in range(len(outcomes)):
    if outcomes[i] == verdicts[i]:
        correct += 1

accuracy = correct/ len(outcomes)
print(accuracy)
        
```
