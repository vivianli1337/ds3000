```python
import sqlite3
import pandas as pd
```

```python
# Make the connection
con = sqlite3.connect('vsx.db')
```

```python
# Load the data
vsx = pd.read_csv('vsx_small.csv')
vsx
```

```python
vsx.to_sql("vsx", con, if_exists='replace') #fail, replace, append
```

```python
con.close()
```
