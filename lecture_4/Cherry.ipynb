```python
# import library
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# open fill
cherry = pd.read_csv('cherry.csv')
cherry
```

```python
# forward fill
cherry['FLOWERING_DOY'] = cherry['FLOWERING_DOY'].fillna(method='ffill')
cherry.dropna(inplace=True) # not gon copy
```

```python
# show
print(list(cherry.FLOWERING_DOY))
```

```python
cherry['MA_10'] = cherry['FLOWERING_DOY'].rolling(window=10).mean()
cherry
```

```python
plt.plot(cherry.YEAR, cherry.FLOWERING_DOY)
```

```python
plt.scatter(cherry.YEAR, cherry.FLOWERING_DOY, color='#FF66B2', s = 10, marker='x')
plt.plot(cherry.YEAR, cherry.MA_10, color='lightblue')
plt.grid()
```
